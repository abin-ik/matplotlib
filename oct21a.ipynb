```python
flight_tf = sns.load_dataset("flights")
flight_tf
```

```python
flight_tf  = flight_tf.pivot_table(index = "month",columns = "year" , values = "passengers")
flight_tf
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(flight_tf,cmap ="Blues",linecolor="Black",linewidth  = 0.5)
plt.yticks(rotation=0)
plt.show()
```

```python
sns.clustermap(flight_tf,cmap = "Blues",standard_scale = 1,annot = True)
```
