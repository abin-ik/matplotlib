```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
bmw= pd.read_csv("bmw.csv")
n =bmw.isnull().sum()
n
```

```python
bmw.head(10)
```

```python
#does increase in tax affect the price of cars for each category
plt.figure(figsize=(11,8))
sns.barplot(x="tax",y="price",hue="transmission",data =bmw)
plt.tight_layout()
```

for automatic cars the tax is lesser compared to the other ones .the price increases when the tax is increased for the semi auo machines

```python
#does the engine size affect the mileage

sns.boxplot(x="engineSize",y="mileage",data= bmw)
```

engine size do effect the mileage.if the engine size is close to 0 and if the engine size is close to 6  the mileage is less.if the engine size is meidan level of 2.5 we get the best mileage 

```python
plt.figure(figsize=(1,6))
sns.barplot(x = "model", y = "price",hue= "transmission",data =bmw)
plt.tight_layout()
```

```python
#which is the model of bmw which got the sals more
plt.figure(figsize=(17,8))
sns.countplot(x ="model",data=bmw)
plt.tight_layout()
```

3 series model is mostly bought

```python
#does the production of automatic and semi automatic cars increases over the years comapared to manual
sns.histplot(x="year",hue="transmission",data = bmw)
```

There was a significant increase in production of automatic and semi automatic cars after 2015

```python
# do the production of electric cars increased
sns.histplot(x="year",hue="fuelType",data = bmw)
```

still the production of diesal car is in demand

```python
plt.figure(figsize=(15,6))
elec1 = bmw[(bmw["transmission"]=="Manual") | (bmw["model"]=="2 Series")]

sns.jointplot(x="year",y="price" ,data=elec1) 
plt.tight_layout()
plt.show()
```
